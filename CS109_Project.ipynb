```python
def count_words(count_words_list, counts_dict):
  for word in count_words_list:
    word = word.strip().lower()
    word = re.sub(r'[^\w\s\']', '', word)
    if word not in counts_dict:
        counts_dict[word] = 1
    else:
        counts_dict[word] += 1
```

```python
def create_df(word_counts):
  df = pd.DataFrame(list(word_counts.items()), columns=['Word', 'Count'])

  s = df["Count"].sum()
  df["Pi"] = df["Count"] / s

  df["Count times log pi"] = df["Count"] * np.log(df["Pi"])

  df.head()
  return df
```

```python
import os
import pandas as pd
import math

def combo(dfs):
  # Concatenate all DataFrames into a single DataFrame
  combined_df = pd.concat(dfs)

  # Group by the 'Word' column and aggregate counts
  grouped_df = combined_df.groupby('Word', as_index=False).agg({'Count': 'sum'})

  # Recompute Pi and Count times log pi
  total_count = grouped_df['Count'].sum()
  grouped_df['Pi'] = grouped_df['Count'] / total_count
  # grouped_df['Count times log pi'] = grouped_df['Count'] * grouped_df['Pi'].apply(lambda x: -1 * (0 if x == 0 else math.log(x)))

  # Sort the DataFrame by 'Word'
  marlowe_agg = grouped_df.sort_values(by='Word')

  # Display the resulting DataFrame
  #print(marlowe_agg)
  #print(marlowe)

  return marlowe_agg
```

```python
import random

def generate_random_numbers(num_values, max_value):
    random_numbers = []
    for _ in range(num_values):
        random_numbers.append(random.randint(0, max_value))
    return random_numbers

# # Example usage
# num_values = 10
# min_value = 1
# max_value = 100
# random_numbers = generate_random_numbers(num_values, min_value, max_value)
# print("Generated random numbers:", random_numbers)
```

```python
# Define the function to calculate the sum of "Count times log pi"
def calc_sum_count_times_log_pi(df):
    return df["Count times log pi"].sum()
```

```python
avg_sh_b = 0
avg_sh_m = 0
avg_sh_v = 0

for i in range(10):
  sh_len = len(shakes)
  # m_len = len(marlowe)
  # v_len = len(vere)
  # b_len = len(bacon)
  # print(sh_len, m_len, v_len, b_len, len(shakes_dfs), len(marlowe_dfs), len(vere_dfs), len(bacon_dfs))
  sh_ind = generate_random_numbers(sh_len//2, sh_len-1)
  # m_ind = generate_random_numbers(m_len//2, m_len-1)
  # v_ind = generate_random_numbers(v_len//2, v_len-1)
  # b_ind = generate_random_numbers(b_len//2, b_len-1)
  # print(sh_ind, m_ind, v_ind, b_ind)
  sh_dfs = [shakes_dfs[idx] for idx in sh_ind]
  # m_dfs = [marlowe_dfs[idx] for idx in m_ind]
  # v_dfs = [vere_dfs[idx] for idx in v_ind]
  # b_dfs = [bacon_dfs[idx] for idx in b_ind]
  sh_other_dfs = [shakes_dfs[idx] for idx in range(len(shakes)) if idx not in  sh_ind]
  # print(sh_dfs)
  sh_combined = combo(sh_dfs)
  sh_other_combined = combo(sh_other_dfs)
  m_combined = combo(marlowe_dfs)
  v_combined = combo(vere_dfs)
  b_combined = combo(bacon_dfs)
  # print(sh_other_dfs)
  # print(sh_combined['Pi'])
  sh_combined['Count times log pi'] = sh_other_combined['Count'] * sh_combined['Pi'].apply(lambda x: -1 * (0 if x == 0 else math.log(x)))
  m_combined['Count times log pi'] = sh_other_combined['Count'] * m_combined['Pi'].apply(lambda x: -1 * (0 if x == 0 else math.log(x)))
  v_combined['Count times log pi'] = sh_other_combined['Count'] * v_combined['Pi'].apply(lambda x: -1 * (0 if x == 0 else math.log(x)))
  b_combined['Count times log pi'] = sh_other_combined['Count'] * b_combined['Pi'].apply(lambda x: -1 * (0 if x == 0 else math.log(x)))

  # Calculate the sum of "Count times log pi" for each combination
  shakespeare_sum = calc_sum_count_times_log_pi(sh_combined)
  # shakespeare_sum2 = calc_sum_count_times_log_pi(shakes2_combined)
  bacon_sum = calc_sum_count_times_log_pi(b_combined)
  marlowe_sum = calc_sum_count_times_log_pi(m_combined)
  vere_sum = calc_sum_count_times_log_pi(v_combined)

  sh_b = shakespeare_sum / bacon_sum
  sh_m = shakespeare_sum / marlowe_sum
  sh_v = shakespeare_sum / vere_sum

  print(f"For iteration {i + 1}:")
  print(f"Shakespeare vs. Authors:")
  print(f"Shakespeare sum / Bacon sum: {sh_b}")
  print(f"Shakespeare sum / Marlowe sum: {sh_m}")
  print(f"Shakespeare sum / Vere sum: {sh_v}")
  print()

  avg_sh_b += sh_b
  avg_sh_m += sh_m
  avg_sh_v += sh_v


avg_sh_b = avg_sh_b / 10
avg_sh_m = avg_sh_m / 10
avg_sh_v = avg_sh_v / 10

print("Average Likelihood Shakespeare wrote those plays rather than Bacon: ", avg_sh_b)
print("Average Likelihood Shakespeare wrote those plays rather than Marlowe: ", avg_sh_m)
print("Average Likelihood Shakespeare wrote those plays rather than Vere: ", avg_sh_v)
```
